```python
## Drawing different shapes on Image
```

```python
### Blank white Image
img = np.zeros((512,512,3),np.uint8)+255
#img[:] = (100,150,200) 
plt.imshow(img)
plt.show()
```

```python
### Draw a line 
img = np.zeros((512,512,3),np.uint8)+255
# image,first,2nd,color,width
cv2.line(img,(20,300),(300,20),(255,0,255),10)
plt.imshow(img)
plt.show()
```

```python
## Draw a rectangle
img = np.zeros((512,512,3),np.uint8)+255
# image,first,2nd,color,width
cv2.rectangle(img,(20,300),(300,20),(0,0,255),10)
plt.imshow(img)
plt.show()
```

```python
## Draw a rectangle
img = np.zeros((512,512,3),np.uint8)+255
# image,first,2nd,color,width
cv2.rectangle(img,(20,300),(300,20),(0,0,255),-1)
plt.imshow(img)
plt.show()
```

```python
## Draw the circle
img = np.zeros((512,512,3),np.uint8)+255
# image,center,radius,color,width
cv2.circle(img,(250,250),200,(0,0,255),5)
cv2.circle(img,(250,250),10,(255,0,0),-1)
plt.imshow(img)
plt.show()
```

```python
### Draw a Polygon
img = np.zeros((512,512,3),np.uint8)+255
points = np.array([[10,200],[100,100],[500,50]])
# image,[points],isclosd,color,width
cv2.polylines(img,[points],True,(0,0,255),5)
plt.imshow(img)
plt.show()
```

```python
## Writing Text On Image
img = np.zeros((512,512,3),np.uint8)+255
# image,text,point,style,size,color,width
cv2.putText(img,"Umang",(100,200),cv2.FONT_HERSHEY_COMPLEX,2,(0,0,255),2)
plt.imshow(img)
plt.show()
```
